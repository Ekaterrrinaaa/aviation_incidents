```python
import pandas as pd
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.estimators import BayesianEstimator # Может понадобиться для state_names, если модель не сохранена с ними
from pgmpy.inference import VariableElimination
import json
import numpy as np
import pickle # Для загрузки/сохранения модели

# --- Константы ---
MODEL_FILE = 'trained_bn_model.pkl' # Файл для сохранения/загрузки обученной модели
CONFIG_FILE = 'bn_config.json'      # Файл конфигурации (нужен для имен исходов)
EVIDENCE_FILE = 'incident_evidence.json'      # Входной файл со свидетельствами

# --- Функция Загрузки Обученной Модели ---
def load_model(filename=MODEL_FILE):
    """Загружает обученную модель из файла."""
    try:
        with open(filename, 'rb') as f:
            model = pickle.load(f)
        print(f"Обученная модель успешно загружена из файла: {filename}")
        # Проверка, что CPT загрузились
        if not model.cpds:
             print("ПРЕДУПРЕЖДЕНИЕ: Модель загружена, но не содержит обученных CPT. Инференс невозможен.")
             return None
        return model
    except FileNotFoundError:
        print(f"Ошибка: Файл с обученной моделью '{filename}' не найден.")
        print("Запустите сначала основной скрипт обучения (BSD.py) с функцией сохранения модели.")
        return None
    except Exception as e:
        print(f"Ошибка при загрузке модели: {e}")
        return None

# --- Функция Загрузки Конфигурации ---
def load_config(filename=CONFIG_FILE):
    """Загружает конфигурацию из JSON."""
    try:
        with open(filename, 'r', encoding='utf-8') as f:
            config = json.load(f)
        print(f"Конфигурация JSON успешно загружена из файла: {filename}")
        return config
    except FileNotFoundError:
        print(f"Ошибка: Файл конфигурации '{filename}' не найден.")
        return None
    except Exception as e:
        print(f"Ошибка при загрузке конфигурации: {e}")
        return None

# --- Функция Загрузки Свидетельств ---
def load_evidence(filename=EVIDENCE_FILE):
    """Загружает свидетельства из JSON."""
    try:
        with open(filename, 'r', encoding='utf-8') as f:
            evidence_data = json.load(f)
        # Проверяем, есть ли ключ 'evidence'
        if 'evidence' not in evidence_data:
             print(f"Ошибка: В файле '{filename}' отсутствует ключ 'evidence'.")
             return None
        evidence_dict = evidence_data['evidence']
        print(f"Свидетельства успешно загружены из файла: {filename}")
        return evidence_dict
    except FileNotFoundError:
        print(f"Ошибка: Файл со свидетельствами '{filename}' не найден.")
        return None
    except json.JSONDecodeError:
         print(f"Ошибка: Файл '{filename}' не является корректным JSON.")
         return None
    except Exception as e:
        print(f"Ошибка при загрузке свидетельств: {e}")
        return None

# --- Функция Предсказания Исхода ---
def predict_outcome(model, evidence, config):
    """Выполняет логический вывод для предсказания исхода."""
    if not model or not model.cpds:
        print("Ошибка: Модель не загружена или не обучена.")
        return None
    if not evidence:
        print("Ошибка: Свидетельства не загружены.")
        return None
    if not config or 'outcome_node' not in config or 'outcome_states' not in config:
        print("Ошибка: Конфигурация не загружена или некорректна.")
        return None

    outcome_node = config['outcome_node']
    outcome_states_config = config['outcome_states']

    print("\n--- Выполнение логического вывода ---")
    print("Заданные свидетельства (Evidence):")
    # Фильтруем свидетельства, оставляя только те узлы, что есть в модели
    valid_evidence = {}
    for node, state in evidence.items():
        if node in model.nodes():
            # Дополнительная проверка: соответствует ли состояние типу/диапазону?
            # (Для простоты пока опустим, предполагаем корректность evidence.json)
            valid_evidence[node] = state
            print(f"  - {node}: {state}")
        else:
            print(f"  - ПРЕДУПРЕЖДЕНИЕ: Узел '{node}' из файла свидетельств отсутствует в модели, игнорируется.")

    if not valid_evidence:
         print("Ошибка: Нет валидных свидетельств для выполнения запроса.")
         return None

    try:
        inference = VariableElimination(model)
        # Запрос вероятностного распределения для узла исхода при заданных свидетельствах
        result_distribution = inference.query(variables=[outcome_node], evidence=valid_evidence)

        print(f"\nПредсказанное распределение вероятностей для '{outcome_node}':")
        print(result_distribution)

        # Вывод вероятностей для каждого состояния исхода
        print("\nВероятности конкретных исходов:")
        results = {}
        for state_info in outcome_states_config:
            code = state_info['code']
            label = state_info.get('label', f'Состояние {code}')
            # Используем get_value для извлечения вероятности конкретного состояния
            try:
                 # Создаем словарь для get_value: {имя_узла_исхода: код_состояния}
                 state_dict = {outcome_node: code}
                 probability = result_distribution.get_value(**state_dict)
                 results[label] = probability
                 print(f"  P({outcome_node}={code} - {label} | Свидетельства) = {probability:.4f}")
            except KeyError:
                 print(f"  ПРЕДУПРЕЖДЕНИЕ: Состояние {code} для '{outcome_node}' отсутствует в результате запроса (вероятность 0).")
                 results[label] = 0.0
            except Exception as e_gv:
                 print(f"  Ошибка при получении значения для состояния {code}: {e_gv}")
                 results[label] = None

        return results # Возвращаем словарь с вероятностями

    except ValueError as e:
        print(f"\nОШИБКА ValueError во время выполнения запроса: {e}")
        print("Возможные причины:")
        print(" - Некорректные значения состояний в файле свидетельств.")
        print(" - Проблемы с обученной моделью (например, нулевые вероятности в CPT).")
    except Exception as e:
        print(f"\nНеожиданная ОШИБКА во время выполнения запроса: {e}")
        return None


# --- Основной блок выполнения ---
if __name__ == "__main__":
    # 1. Загружаем конфигурацию
    config_data = load_config()
    if not config_data:
        exit()

    # 2. Загружаем обученную модель
    trained_model = load_model()
    if not trained_model:
        # --- Если модель не найдена, пробуем обучить ее заново ---
        # --- СКОПИРУЙТЕ СЮДА ВЕСЬ КОД ОБУЧЕНИЯ ИЗ ВАШЕГО BSD.py ---
        # --- от загрузки данных до model.fit() и сохранения модели ---
        print("\nПопытка обучить модель заново, так как файл не найден...")

        # --- Начало скопированного кода из BSD.py ---
        try:
            df = pd.read_csv('aviation_incidents_refined.csv') # Загружаем УТОЧНЕННЫЕ данные
            print(f"Данные CSV для обучения ({df.shape[0]} строк) загружены.")
        except FileNotFoundError:
            print("Ошибка: Файл 'aviation_incidents_refined.csv' для обучения не найден.")
            exit()
        except Exception as e:
             print(f"Ошибка при загрузке данных для обучения: {e}")
             exit()

        # ... (Весь код подготовки данных df_prepared из BSD.py) ...
        # ... (Определение структуры edges и model_nodes_list из BSD.py) ...
        # ... (Проверки кардинальностей и на отрицательные значения) ...

        # Создание модели (скопировано из BSD.py)
        factor_nodes = [f['id'] for f in config_data['factors']]
        outcome_node = config_data['outcome_node']
        final_factor_nodes = [node for node in factor_nodes if node in df_prepared.columns]
        if outcome_node not in df_prepared.columns:
             print("Критическая ошибка: outcome_node отсутствует в df_prepared при переобучении")
             exit()
        edges = [] # Пересоздайте ваш список edges здесь (скопируйте из BSD.py)
        # ... (код edges.extend(...) из вашего BSD.py) ...
        unique_edges = list(set(edges))
        model_nodes_list = [node for node in (final_factor_nodes + [outcome_node]) if node in df_prepared.columns]
        trained_model = DiscreteBayesianNetwork()
        trained_model.add_nodes_from(model_nodes_list)
        trained_model.add_edges_from(unique_edges)
        if not nx.is_directed_acyclic_graph(trained_model):
             print("Ошибка: Структура сети при переобучении содержит циклы!")
             exit()
        print("Структура для переобучения создана.")

        # Подготовка state_names (скопировано из BSD.py)
        state_names = {}
        for factor in config_data['factors']:
             if factor['id'] in model_nodes_list:
                  state_names[factor['id']] = sorted([s['code'] for s in factor['states']])
        if outcome_node in model_nodes_list:
            state_names[outcome_node] = sorted([s['code'] for s in config_data['outcome_states']])
        print("State_names для переобучения подготовлен.")

        # Обучение (скопировано из BSD.py)
        data_to_fit = df_prepared[model_nodes_list].copy()
        try:
            print("Начало переобучения модели...")
            trained_model.fit(data=data_to_fit, estimator=BayesianEstimator, state_names=state_names, equivalent_sample_size=10)
            print("Модель успешно переобучена.")
            # Сохраняем свежеобученную модель
            try:
                with open(MODEL_FILE, 'wb') as f:
                    pickle.dump(trained_model, f)
                print(f"Обученная модель сохранена в файл: {MODEL_FILE}")
            except Exception as e_save:
                print(f"Ошибка при сохранении переобученной модели: {e_save}")

        except Exception as e_fit:
            print(f"Ошибка во время переобучения модели: {e_fit}")
            trained_model = None # Сбрасываем модель, если обучение не удалось
            exit()
        # --- Конец скопированного кода из BSD.py ---

    if not trained_model or not trained_model.cpds:
        print("Не удалось загрузить или обучить модель. Выход.")
        exit()

    # 3. Загружаем свидетельства
    evidence_dict = load_evidence()
    if not evidence_dict:
        exit()

    # 4. Делаем предсказание
    predicted_probabilities = predict_outcome(trained_model, evidence_dict, config_data)

    # 5. Опционально: Выводим наиболее вероятный исход
    if predicted_probabilities:
        most_likely_outcome = max(predicted_probabilities, key=predicted_probabilities.get)
        print(f"\nНаиболее вероятный исход: {most_likely_outcome} (Вероятность: {predicted_probabilities[most_likely_outcome]:.4f})")
```
